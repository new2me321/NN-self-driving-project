```python
from PIL import Image
from PIL import ImageOps
import os
import os.path

filepath = '/home/karina/projects/mycar/data/tub_92_22-05-29/images/'
border = (140, 80, 0, 0) # left, top, right, bottom

for filename in os.listdir(filepath):
    image = Image.open(os.path.join(filepath, filename))
    ImageOps.crop(image, border).save(os.path.join(filepath, filename))
```
